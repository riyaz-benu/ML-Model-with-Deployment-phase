```python
import pandas as pd
dataset=pd.read_csv('50_Startups.csv')
dataset=pd.get_dummies(dataset,drop_first=True)
indep=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dep=dataset[['Profit']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python
sc_y=StandardScaler()
y_train=sc_y.fit_transform(y_train)
y_test=sc_y.transform(y_test)
```

```python
import pickle
with open('sc_x.pkl', 'wb') as f:
    pickle.dump(sc_x, f)

with open('sc_y.pkl', 'wb') as f:
    pickle.dump(sc_y, f)
```

```python
from sklearn.svm import SVR
regressor_xy=SVR(kernel="linear",C=10)
regressor_xy.fit(X_train,y_train)
```

```python
with open('regressor_xy.pkl', 'wb') as f:
    pickle.dump(regressor_xy, f)
```

```python
y_pred=regressor_xy.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
with open('sc_x.pkl', 'rb') as f:
    loaded_scaler_x = pickle.load(f)

with open('sc_y.pkl', 'rb') as f:
    loaded_scaler_y = pickle.load(f)

with open('regressor_xy.pkl', 'rb') as f:
    loaded_model_xy = pickle.load(f)

# Transform test data using the loaded scaler
preinput = loaded_scaler_x.transform([[1289,3455, 7652,0,1]])

# Predict with the loaded model
predictions = loaded_model_xy.predict(preinput)

predictions
```
